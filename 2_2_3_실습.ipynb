ggplot으로 그래프 그리기 review

```python
#패키지 import
from plotnine import * #ggplot 패키지
import matplotlib.pyplot as plt #시각화 패키지
import seaborn as sns #예시 데이터 패키지
flights = sns.load_dataset('flights') #예시 데이터
```

```python
#데이터 확인
flights.head()
```

```python
#histogram
ggplot(flights, aes(x='passengers')) + geom_histogram()
```

```python
#bar plot
ggplot(flights, aes(x='month', y='passengers')) + geom_bar(stat='identity') 
```

```python
#box plot
ggplot(flights, aes(x='month', y='passengers')) + geom_boxplot() 
```

```python
#scatter plot
ggplot(flights, aes(x='year', y='passengers')) + geom_point() 
```

```python
#facet_grid: 변수를 그룹화하여 시각화(month로 그룹)
ggplot(flights, aes(x='year',y='passengers')) + geom_point() + facet_grid('month ~.')
```

```python
#구글 마운트 안되는 경우 파일 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/hyeokkukim/reading_bigdata/main/2-2-3_%E1%84%89%E1%85%B5%E1%86%AF%E1%84%89%E1%85%B3%E1%86%B8.csv',encoding='cp949')
```

```python
#데이터 확인
df.head()
```

```python
#데이터 구조 확인
df.describe()
```

```python
#colab용 한글폰트 깨짐 방지위해 컬럼명 영어로 변경
df.columns = ['key_number','key_code','senior','age','owner','income']
df.head()
```

```python
#가구주 연령 히스토그램
ggplot(df, aes(x='age')) + geom_histogram()
```

```python
#노인가구의 소득 plot
ggplot(df, aes(x='senior', y='income')) + geom_bar(stat='identity')
```

```python
#연령별 소득 scatter plot
ggplot(df, aes(x='age', y='income')) + geom_point()
```

```python
#가구별 소득 평균 plot
ggplot(df.groupby(['key_number'])['income'].mean().reset_index(),aes(x='key_number',y='income'))+geom_bar(stat='identity')
```
